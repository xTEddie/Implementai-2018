```python
def get_x(p):
    df_ = pd.read_csv(p)
    df = df_.fillna(0).rename(columns={'Unnamed: 0' : 'frame_id'})
    df_x = pd.DataFrame()
    for i in (range(0, max(df.frame_id))[::15]):
        temp_timestamp = df.loc[df['frame_id'] == i].values[:,1:-1]
        temp_array = np.amax(np.absolute(temp_timestamp), axis = 0)
        df_x = df_x.append(pd.DataFrame( temp_array.reshape(-1,len(temp_array))))
    return np.array(df_x)
```

```python
def train():
    X = get_x(dir_path + "/bad_5_velocity.csv")
    y = (long_labels.values)[4].T
    clf = DecisionTreeClassifier(random_state=0)
    clf.fit(X, y)
    return clf
```

```python
model = train()
```

```python
def predict_labels(clf, p, n=4):
    X = get_x(dir_path + "/" + p)
    print(p)
    print(X.shape)
    y_pred = model.predict(X, check_input=True)
    print(y_pred.shape)
    y_true = (long_labels.values)[n]
    print(y_true.shape)

    accuracy = accuracy_score(y_true, y_pred)
    return y_pred, accuracy
```
